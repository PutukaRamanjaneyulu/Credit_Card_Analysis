```python
# drop function will be Drop the Coloumns Mention to the Axis=1 -> is Columns axis=0 -> Rows
df=df.drop('MonthlyIncome.1',axis=1)
df
```

```python
# After removing the Columns information
df.info()
```

```python
df.isnull().sum()
```

```python
# Drops the Enter Row when single Column is null Value
df.dropna(subset='Gender',inplace=  True,axis=0)
```

```python
df.isnull().sum()
```

```python
# MonthlyIncome Mean
df['MonthlyIncome'].mean()
```

```python
#null Values Are Fill The MonthlyIncome mean
df['MonthlyIncome']=df['MonthlyIncome'].fillna(df['MonthlyIncome'].mean())
```

```python
df
```

```python
# NumberofDepents Finding the Mode
df['NumberOfDependents'].mode()[0]
```

```python
#NumberofDepents null values are Filling to the Mode Function
df['NumberOfDependents']=df['NumberOfDependents'].fillna(df['NumberOfDependents'].mode()[0])
```
